```python
########

def schoolid(row):
    if row['school'] == 'ironhack':
        return 10828
    elif row['school'] == 'app-academy':
        return 10525
    elif row['school'] == 'springboard':
        return 11035
    elif row['school'] == 'udacity':
        return 11118
    elif row['school'] == 'brainstation':
        return 10571
    elif row['school'] == 'practicum-bootcamp':
        return 11225
```

```python
comments.apply(schoolid, axis=1)
```

```python
comments['school_id'] = comments.apply(schoolid, axis=1)
```

```python
## We send the CLEAN data back to SQL -- project2v2

from sqlalchemy import create_engine

# Credentials to database connection
hostname='127.0.0.1'
dbname="project2v2"
uname="root"
pwd="PASSWORD"

# Create dataframe
comments
courses
locations
schools


# Create SQLAlchemy engine to connect to MySQL Database
engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}"
				.format(host=hostname, db=dbname, user=uname, pw=pwd))

# Convert dataframe to sql table                                   
comments.to_sql('comments', engine, index=False)
courses.to_sql('courses', engine, index=False)
locations.to_sql('locations', engine, index=False)
schools.to_sql('schools', engine, index=False)
```
