```python
### DROPPING COLUMNS in Comments

comments.drop('name', inplace=True, axis=1)
comments.drop('tagline', inplace=True, axis=1)
comments.drop('body', inplace=True, axis=1)
comments.drop('rawBody', inplace=True, axis=1)
comments.drop('createdAt', inplace=True, axis=1)
comments.drop('queryDate', inplace=True, axis=1)


### in School

schools.drop('website', inplace=True, axis=1)
schools.drop('description', inplace=True, axis=1)
schools.drop('LogoUrl', inplace=True, axis=1)


## in Location

locations.drop('country.id', inplace=True, axis=1)
locations.drop('country.name', inplace=True, axis=1)
locations.drop('city.id', inplace=True, axis=1)
locations.drop('city.keyword', inplace=True, axis=1)
locations.drop('state.id', inplace=True, axis=1)
locations.drop('state.name', inplace=True, axis=1)
locations.drop('state.abbrev', inplace=True, axis=1)
locations.drop('state.keyword', inplace=True, axis=1)

locations
```

```python
## We send the CLEAN data back to SQL -- project2v2

from sqlalchemy import create_engine

# Credentials to database connection
hostname='127.0.0.1'
dbname="project2v2"
uname="root"
pwd="PASSWORD"

# Create dataframe
comments
courses
locations
schools


# Create SQLAlchemy engine to connect to MySQL Database
engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}"
				.format(host=hostname, db=dbname, user=uname, pw=pwd))

# Convert dataframe to sql table                                   
comments.to_sql('comments', engine, index=False)
courses.to_sql('courses', engine, index=False)
locations.to_sql('locations', engine, index=False)
schools.to_sql('schools', engine, index=False)
```
